```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
surface1 = pd.read_csv('surface1.csv')
surface2 = pd.read_csv('surface2.csv')
surface4 = pd.read_csv('surface4.csv')
surface7 = pd.read_csv('surface7.csv')
```

```python
surface1.drop('Unnamed: 0', axis=1)
surface2.drop('Unnamed: 0', axis=1)
surface4.drop('Unnamed: 0', axis=1)
surface7.drop('Unnamed: 0', axis=1);
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.view_init(elev=20., azim=-135)
ax.scatter3D(surface1['x'], surface1['y'], np.asarray(surface1['z']), c=np.asarray(surface1['cs']), cmap=cm.plasma,
                       linewidth=0, antialiased=False, alpha=.75)
ax.margins(0)
plt.tight_layout()
plt.savefig('./plots/fullbox_py.png',format='png',dpi=200)
plt.show()
```
